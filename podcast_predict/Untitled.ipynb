```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("train.csv")
data.head()
```

```python
data.columns
```

```python
data=data.drop(columns=["id"])
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.fillna(0, inplace=True)
```

```python
data["Episode_Sentiment"]=data["Episode_Sentiment"].map({"Positive":1,"Negative":0,"Neutral":1})
```

```python
X=data[["Episode_Length_minutes","Host_Popularity_percentage","Guest_Popularity_percentage","Episode_Sentiment"]]
```

```python
y=data[["Listening_Time_minutes"]]
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
dataset=pd.read_csv("test.csv")
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset=dataset.drop(columns=["id","Podcast_Name","Episode_Title","Genre","Publication_Day","Publication_Time","Number_of_Ads"])
```

```python
dataset["Episode_Sentiment"]=dataset["Episode_Sentiment"].map({"Positive":1,"Negative":0,"Neutral":1})
```

```python
dataset.fillna(0, inplace=True)
```

```python
dataset.isna().sum()
```

```python
y_pred = model.predict(dataset)
```

```python
print(y_pred)
```

```python
dataset["listening_Time_minutes"]=y_pred
```
